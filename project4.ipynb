```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {  
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],  
    'Avg_Temp': [10, 12, 15, 18],  
    'Rainfall': [80, 50, 40, 30]  
}  

plt.plot(data["Month"],data["Avg_Temp"],marker="h",linestyle="-",color="red")
plt.title("A line chart for temperature across Months",fontsize=16,fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Average Tempreature")
plt.show()
```

```python
plt.bar(data["Month"],data["Rainfall"],color=["red","green","blue","yellow"])
plt.title("A bar chart for rainfall across months",fontsize=18,fontweight='bold')
plt.xlabel("Months")
plt.ylabel("Rainfall")
plt.show()
```

```python
plt.scatter(data["Avg_Temp"],data["Rainfall"],color="green",marker="p")
plt.title('Relationship between Temprature and Rainfall',fontsize=18,fontweight="bold",color="navy")
plt.xlabel("Temprature",fontsize=10)
plt.ylabel("Rainfall",fontsize=10)
plt.grid(True,linestyle="-",linewidth=0.7)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

There is inverse relationship between temprature and rainfall.Whenever temprature increase rainfall decreases and vice-versa.
